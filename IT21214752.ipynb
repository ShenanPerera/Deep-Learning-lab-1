```python
plt.hist(np.random.exponential(1,[100000,1]), density=True, bins=100, histtype="step", color="blue",
label="exponential")
plt.hist(np.random.rand(100000), density=True, bins=100, histtype="step", color="green",
label="uniform")
plt.hist(np.random.randn(100000), density=True, bins=100, histtype="step", color="red",
label="normal")
plt.axis([-4, 10, 0, 1.1])
plt.legend(loc = "upper left")
plt.title("Random exponential distributions")
plt.xlabel("Value")
plt.ylabel("Density")
plt.show()
```

```python
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline
import seaborn as sns

!wget -q https://elitedatascience.com/wp-content/uploads/2022/07/Pokemon.csv
df = pd.read_csv('Pokemon.csv', index_col=0, encoding='latin')
df
```

```python
# pearson
stats_df = df.drop(['Name','Type 1','Type 2','Total', 'Stage', 'Legendary'], axis=1)
corr = stats_df.corr(method = 'pearson')
plt.figure(figsize=(9,8))
sns.heatmap(corr)
```

```python
stats_df = df.drop(['Name','Type 1','Type 2','Total', 'Stage', 'Legendary'], axis=1)
spearman_corr= stats_df.corr(method = 'spearman')
plt.figure(figsize=(9,8))
sns.heatmap(spearman_corr)
```
